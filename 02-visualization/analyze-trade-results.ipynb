```python
import lzma
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# describe columns type
dtype = {
    'symbol': 'category',
    'timeframe': 'category',
    'type': 'category',
}
parse_dates = ['open_time', 'close_time']

file = lzma.open('data/trade_results.csv.xz')
trade_results = pd.read_csv(file, dtype=dtype, parse_dates=parse_dates)
trade_results.head()
```

# Предобработка данных

```python
# Drop gold and silver trades
trade_results = trade_results[~trade_results.symbol.str.match('XAGUSD|XAUUSD')].reset_index(drop=True)
# Update categories
categories = trade_results['symbol'].unique()
trade_results['symbol'].cat.set_categories(categories, inplace=True)
```

```python
is_win = lambda x: x > 0
trade_results.loc[:, 'is_win'] = trade_results['profit_in_pips'].transform(is_win)
trade_results.loc[:, 'result_in_pips'] = trade_results['profit_in_pips'].transform(np.abs)
# trade_results['profit_in_pips'].apply((lambda x: x > 0), axis=1)
trade_results.head()
```

# Проанализировать результаты торговли

## Прибыль или убыток

```python
# grouped = trade_results.groupby(['symbol', 'is_win']).agg({'result_in_pips': ['sum']})
grouped = trade_results.groupby(['symbol', 'is_win']).agg({'result_in_pips': ['sum', 'count']})
grouped.head(4)
```

### Отобразим столбчатые диаграммы для сравнения прибыли/убытков
Диаграммы в разрезе по валютным парам:
* Прибыль и убыток в пипсах:
    * абсолютные значения
    * относительные значения
* Количество прибыльных и убыточных сделок:
    * абсолютные значения
    * относительные значения

```python
from pylab import rcParams


rcParams['figure.figsize'] = 15,10
# plt.style.use(['dark_background', 'presentation'])
# plt.style.use(['seaborn-darkgrid'])
# plt.style.use(['seaborn-notebook'])
# plt.style.use(['dark_background'])

# подготавливам значения для X
symbols = trade_results['symbol'].unique()

symbols_win = [(symbol, is_win) for symbol, is_win in grouped.index.values if is_win]
sum_win = grouped.loc[symbols_win, ('result_in_pips', 'sum')] / 1000
count_win = grouped.loc[symbols_win, ('result_in_pips', 'count')]

symbols_loss = [(symbol, is_win) for symbol, is_win in grouped.index.values if not is_win]
sum_loss = grouped.loc[symbols_loss, ('result_in_pips', 'sum')] / 1000
count_loss = grouped.loc[symbols_loss, ('result_in_pips', 'count')]


# print(symbols_win)
fig, axs = plt.subplots(2, 2, constrained_layout=True)
ax1 = axs[0, 0]
ax1.bar(symbols, sum_loss, label='убыток')
ax1.bar(symbols, sum_win, label='прибыль')

ax1.xaxis.set_tick_params(rotation=90)
ax1.set_xlabel('Пары')
ax1.set_ylabel('Пипсы, тыс.')
ax1.title.set_text('Сопоставление прибыли и убытка')
ax1.legend()

ax2 = axs[0, 1]
ax2.bar(symbols, count_loss, label='убыток')
ax2.bar(symbols, count_win, label='прибыль')

ax2.xaxis.set_tick_params(rotation=90)
ax2.set_ylabel('Сделки, шт.')
ax2.set_xlabel('Пары')
ax2.title.set_text('Прибыльные и убыточные сделки')
ax2.legend()


normalized_sum_win =  (100 * sum_win.values / (sum_win.values + sum_loss.values)).round(0)
normalized_sum_loss =  (100 * sum_loss.values / (sum_win.values + sum_loss.values)).round(0)

normalized_count_win =  (100 * count_win.values / (count_win.values + count_loss.values)).round(0)
normalized_count_loss =  (100 * count_loss.values / (count_win.values + count_loss.values)).round(0)
# print(sum_win.values)
# print(sum_loss)
# print(normalized_sum_win)
# print(normalized_sum_loss)

ax3 = axs[1, 0]
p = ax3.bar(symbols, normalized_sum_loss, label='убыток')
ax3.bar(symbols, normalized_sum_win, bottom=normalized_sum_loss, label='прибыль')

ax3.bar_label(p, label_type='center', rotation=90, color='white')

ax3.xaxis.set_tick_params(rotation=90)
ax3.set_ylabel('Прибыль/убыток, %')
ax3.set_xlabel('Пары')
ax3.title.set_text('Сопоставление прибыли и убытка, %')
ax3.legend()

ax4 = axs[1, 1]
p = ax4.bar(symbols, normalized_count_loss, label='убыток')
ax4.bar(symbols, normalized_count_win, bottom=normalized_count_loss, label='прибыль')

ax4.bar_label(p, label_type='center', rotation=90, color='white')

ax4.xaxis.set_tick_params(rotation=90)
ax4.set_ylabel('Прибыльные/убыточные сделки, %')
ax4.set_xlabel('Пары')
ax4.title.set_text('Прибыльные/убыточные сделки, %')
ax4.legend()


plt.show()
```

#### Выводы:
* По всем инструментам убыток :(
     * `USDJPY` и `EURUSD` имеют наименьший убыток по сравнению с остальными
* Соотношение прибыльных и убыточных сделок "ужасает" :(

## Взаимосвязь между прибылью и убытком

```python
win = trade_results['is_win']
loss = ~trade_results['is_win']

# reference https://dev.to/thalesbruno/subplotting-with-matplotlib-and-seaborn-5ei8
fig, axs = plt.subplots(2, 1, figsize=[15, 15], constrained_layout=True)
ax1 = axs[0]
ax2 = axs[1]

sns.boxplot(ax=ax1, y=trade_results[win]['result_in_pips'], x=trade_results[win]['symbol'])
sns.boxplot(ax=ax2, x=trade_results[loss]['result_in_pips'], y=trade_results[loss]['symbol'])

ax1.set_title('Прибыльные сделки')
ax1.set_ylabel('Пипсы')
ax1.set_xlabel('Пары')
ax1.xaxis.set_tick_params(rotation=90)

ax2.set_title('Убыточные сделки')
ax2.set_xlabel('Пипсы')
ax2.set_ylabel('Пары')

plt.show()
```

#### Выводы:
* Большинство прибыльных сделок более прибыльны для `GBPCAD`, `GBPAUD`, ...
* Большинство убыточных сделок более убыточны для `GBPCAD`, `GBPAUD`, ... 

На лицо зависимость между ожидаемой (медианной) прибылью и убытком. 

# Проанализируем самую "прибыльную" валютную пару: USDJPY
Проверим влияние временных характеристик открытия сделки на результат торговли с целью найти фильтр для сделок. 

Если найдем подходящий фильтр для `USDJPY`, то попробуем применить его и для других валютных пар.

## Подготовим данные

```python
usdjpy = trade_results[trade_results['symbol'] == 'USDJPY'].reset_index(drop=True)
# drop unused attributes
del usdjpy['open_price']
del usdjpy['close_time']
del usdjpy['close_price']
# add need attributes
usdjpy['year'] = usdjpy['open_time'].dt.year
usdjpy['month'] = usdjpy['open_time'].dt.month
usdjpy['dayofweek'] = usdjpy['open_time'].dt.dayofweek
usdjpy['hour'] = usdjpy['open_time'].dt.hour
usdjpy['minute'] = usdjpy['open_time'].dt.minute
usdjpy.head()
```

## Найдем самые неудачные моменты времени в координатах (час, день недели)

```python
loss = ~usdjpy['is_win']
usdjpy_dayofweek_per_hour = usdjpy[loss].pivot_table(index='hour', columns='dayofweek', values='result_in_pips', aggfunc=len)
sns.heatmap(usdjpy_dayofweek_per_hour.fillna(0), linewidth=0.5)
```

#### Выводы
Наибольшее количество неудачных сделок было открыто:

hour | dayofweek
--- | ---
13 | 4
17 | 2
14 | 2
8 | 4

## Попробуем применить фильтр для USDJPY

```python
f1 = (usdjpy['hour'] == 13) & (usdjpy['dayofweek'] == 4)
f2 = (usdjpy['hour'] == 17) & (usdjpy['dayofweek'] == 2)
f3 = (usdjpy['hour'] == 14) & (usdjpy['dayofweek'] == 2)
f4 = (usdjpy['hour'] == 8) & (usdjpy['dayofweek'] == 4)
skip_loss = ~(f1 | f2 | f3 | f4)

filtered_result = usdjpy[skip_loss]['profit_in_pips'].sum()
unfiltered_result = usdjpy['profit_in_pips'].sum()

print(filtered_result, unfiltered_result)
```

#### Вывод
На 2500 пипсов улучшили результат, но он все равно печален :(

## Посмотрим на результаты применения фильтра для всех пар

### Обогатим данные

```python
trade_results_enriched = trade_results.copy()
# drop unused attributes
del trade_results_enriched['open_price']
del trade_results_enriched['close_time']
del trade_results_enriched['close_price']
# add need attributes
trade_results_enriched['year'] = trade_results_enriched['open_time'].dt.year
trade_results_enriched['month'] = trade_results_enriched['open_time'].dt.month
trade_results_enriched['dayofweek'] = trade_results_enriched['open_time'].dt.dayofweek
trade_results_enriched['hour'] = trade_results_enriched['open_time'].dt.hour
trade_results_enriched['minute'] = trade_results_enriched['open_time'].dt.minute
trade_results_enriched.head()
```

### Подготовим данные для оценки фильтрации

```python
f1 = (trade_results_enriched['hour'] == 13) & (trade_results_enriched['dayofweek'] == 4)
f2 = (trade_results_enriched['hour'] == 17) & (trade_results_enriched['dayofweek'] == 2)
f3 = (trade_results_enriched['hour'] == 14) & (trade_results_enriched['dayofweek'] == 2)
f4 = (trade_results_enriched['hour'] == 8) & (trade_results_enriched['dayofweek'] == 4)
skip_loss = ~(f1 | f2 | f3 | f4)

unfiltered = trade_results_enriched.groupby('symbol').agg({'profit_in_pips': 'sum'})
filtered = trade_results_enriched[skip_loss].groupby('symbol').agg({'profit_in_pips': 'sum'})

unfiltered['filtered?'] = 'no'
filtered['filtered?'] = 'yes'

impact_filter = pd.concat([filtered, unfiltered]).sort_values('profit_in_pips', ascending=False).reset_index()

# Update categories for changing sort order in visualization
categories = impact_filter['symbol'].unique()
impact_filter['symbol'].cat.set_categories(categories, inplace=True)

impact_filter.head(2)
```

```python
sns.barplot(y=impact_filter['symbol'], x=impact_filter['profit_in_pips'], hue=impact_filter['filtered?'])
plt.show()
```

#### Выводы
Видим, что фильтр найденный для `USDJPY` положительно подействовал на все пары.
